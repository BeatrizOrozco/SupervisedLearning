```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('USArrests.csv',usecols=[1,2,3,4])
states=np.squeeze(np.array(pd.read_csv('USArrests.csv',usecols=[0])))
```

```python
scaled_data = StandardScaler().fit(df).transform(df)
pca = PCA(n_components = 2).fit(scaled_data)
x_pca = pca.transform(scaled_data)
components = pd.DataFrame(pca.components_.T, index=df.columns, columns=['PCA1','PCA2'])
```

```python
fig, ax = plt.subplots(figsize=(10,10))
#ax.plot(pc1, pc2,'.')
ax.hlines(0,-4,4, linestyles='dotted', colors='grey')
ax.vlines(0,-4,4, linestyles='dotted', colors='grey')
ax.set_ylim(-2.5,2);
ax.set_xlim(-3.5,3.5);
offset = 1.07
for zz,yy,txt in np.broadcast(x_pca[:,0],-x_pca[:,1],states):
    ax.annotate(txt, (zz,yy))
ax.arrow(0, 0, 2.2*components['PCA1'][0], 2.2*-components['PCA2'][0],color='orange',head_width=0.05, head_length=0.1)
ax.arrow(0, 0, 2*components['PCA1'][1], 2*-components['PCA2'][1],color='orange',head_width=0.05, head_length=0.1)
ax.arrow(0, 0, 2*components['PCA1'][2], 2*-components['PCA2'][2],color='orange',head_width=0.05, head_length=0.1)
ax.arrow(0, 0, 2*components['PCA1'][3], 2*-components['PCA2'][3],color='orange',head_width=0.05, head_length=0.1)
ax.annotate(components.index[0], (2.2*components['PCA1'][0]*offset, 2.2*-components['PCA2'][0]*offset), color='orange')
ax.annotate(components.index[1], (2*components['PCA1'][1]*offset, 2*-components['PCA2'][1]*offset), color='orange')
ax.annotate(components.index[2], (2*components['PCA1'][2]*offset, 2*-components['PCA2'][2]*offset), color='orange')
ax.annotate(components.index[3], (2*components['PCA1'][3]*offset, 2*-components['PCA2'][3]*offset), color='orange')
ax.set_xlabel('First Principal Component')
ax.set_ylabel('Second Principal Component')
```

```python
scaled_data = StandardScaler().fit(df).transform(df)
pca = PCA(n_components = 4).fit(scaled_data)
```

```python
exp_var=pca.explained_variance_ratio_
cum_var=np.cumsum(exp_var)
```

```python
fig,(ax0, ax1) = plt.subplots(ncols=2,figsize=(11, 5))
ax0.plot(np.arange(1,5),exp_var,linewidth=1)
ax0.scatter(np.arange(1,5),exp_var,facecolors='none',edgecolors='blue')
ax0.set_ylabel('Prop. Variance Explained')
ax0.set_xlabel('Principal Component')
ax0.set_ylim(0,1.2)
ax1.plot(np.arange(1,5),cum_var,linewidth=1)
ax1.scatter(np.arange(1,5),cum_var,facecolors='none',edgecolors='blue')
ax1.set_ylabel('Cumulative Prop. Variance Explained')
ax1.set_xlabel('Principal Component')
ax1.set_ylim(0,1.2)
```

## Segunda Parte

```python
df=pd.read_csv('Cars93.csv',usecols=[8,11,12,13,14,15,17,19,20,21,22,23,24,25])
df= df.dropna()
```

```python
var=[]
for i in range(len(np.arange(1,5))):
    scaled_data = StandardScaler().fit(df).transform(df)
    pca = PCA(n_components = i).fit(scaled_data)
    x_pca = pca.transform(scaled_data)
    var.append(sum(pca.explained_variance_ratio_))
#components = pd.DataFrame(pca.components_.T, index=df.columns, columns=['PCA1','PCA2'])
```

```python
var
```

3 componentes explican el 84% de la varianza

```python
plt.plot(np.arange(1,5),var,linewidth=1)
plt.scatter(np.arange(1,5),var,facecolors='none',edgecolors='blue')
plt.hlines(0.8,0,4, linestyles='dotted', colors='grey')
plt.xlabel('No. de Componentes')
plt.ylabel('Varianza')
```

```python
scaled_data = StandardScaler().fit(df).transform(df)
pca = PCA(n_components = 2).fit(scaled_data)
x_pca = pca.transform(scaled_data)
components = pd.DataFrame(pca.components_.T, index=df.columns, columns=['PCA1','PCA2'])


fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
for i in range(len(np.arange(0,14))):
    ax.arrow(0, 0, 3*components['PCA1'][i], 2.5*components['PCA2'][i],color='orange',head_width=0.05, head_length=0.1)
    ax.annotate(components.index[i], (3*components['PCA1'][i]*offset, 2.5*components['PCA2'][i]*offset), color='orange')
```
