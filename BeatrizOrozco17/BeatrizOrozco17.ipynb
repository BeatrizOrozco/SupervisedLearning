# Matriz de confusión

```python
all_train, all_test= train_test_split(data, test_size=0.25)
```

```python
default = list(all_train['default'])
stud = list(all_train['student'])
income = list(all_train['income'])
balance = list(all_train['balance'])
student=[]
for i in range(len(stud)):
    if stud[i]=='Yes':
        student.append(1)
    if stud[i]=='No':
        student.append(0)
X_train=(np.array([student,income,balance])).T
y_train=default
```

```python
default = list(all_test['default'])
stud = list(all_test['student'])
income = list(all_test['income'])
balance = list(all_test['balance'])
student=[]
for i in range(len(stud)):
    if stud[i]=='Yes':
        student.append(1)
    if stud[i]=='No':
        student.append(0)
X_test=(np.array([student,income,balance])).T
y_test=default
```

## Matriz de Confusión para datos training

```python
clf = linear_model.LogisticRegression()
clf.fit(X_train, y_train)
predict_train=clf.predict(X_train)
```

```python
nn_train=0
pp_train=0
np_train=0
pn_train=0
for i in range(len(X_train)):
    if y_train[i]=='No' and predict_train[i]=='No':
        nn_train+=1
    if y_train[i]=='Yes' and predict_train[i]=='Yes':
        pp_train+=1
    if y_train[i]=='No' and predict_train[i]=='Yes':
        np_train+=1
    if y_train[i]=='Yes' and predict_train[i]=='No':
        pn_train+=1
```

```python
m_confusion_train=np.array([[nn_train,np_train],[pn_train,pp_train]])
m_confusion_train
```

## Matriz de Confusión para datos test

```python
clf = linear_model.LogisticRegression(C=1e5, solver='lbfgs')
clf.fit(X_test,y_test)
predict_test=clf.predict(X_test)
```

```python
nn_test=0
pp_test=0
np_test=0
pn_test=0
for i in range(len(X_test)):
    if y_test[i]=='No' and predict_test[i]=='No':
        nn_test+=1
    if y_test[i]=='Yes' and predict_test[i]=='Yes':
        pp_test+=1
    if y_test[i]=='No' and predict_test[i]=='Yes':
        np_test+=1
    if y_test[i]=='Yes' and predict_test[i]=='No':
        pn_test+=1
```

```python
m_confusion_test=np.array([[nn_test,np_test],[pn_test,pp_test]])
m_confusion_test
```
